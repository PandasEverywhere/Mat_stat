```python
import pandas as pd
import numpy as np
```

```python
# Считывание данных.
data = pd.read_csv("6.csv.xls", header=None)
lamb = 105
t_0 = 500
t = 110000
```

```python
# Посчитаем разницы между соседними.
tmp = []
for i in range(len(data)-1):
    tmp += [float(data.ix[i+1] - data.ix[i])]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%pylab inline
pylab.figure(figsize=(10, 6))
plt.hist(tmp, bins=200)
plt.show()
```

Странный пуассон.

E(N_t|N_s) = E(N_t - N_s|N_s) + E(N_s|N_s) = lambda * (t - s) + N_s

```python
# Предсказывание в каждый момент времени t_0 * i.
preds = []
for i in range(t / t_0):
    preds += [float( 1.0 * lamb * (t - (i * t_0))  +   data.ix[i * 1000 / (t / t_0)])]
```

```python
plt.plot(range(220),preds)
```
