```python
norm.ppf(0.025, 0, 0.127)
```

```python
# Байесовская оценка
def bayes(sample, x1, x2):
    return (sum(sample) + (x1/x2**2))/(len(sample) + (1.0/(x2**2)))
```

```python
x = linspace(-10,10,100)
plt.plot(x, cauchy.pdf(x))
plt.show()
```

```python
# Оценка максимального правдоподобия
arr = []
for i in range(1, 200):
    arr += [np.abs(sample[:i].mean())]

# Байесовская оценки
arr1 = []
for i in range(1, 200):
    arr1 += [np.abs(bayes(sample[:i], 0, 0.127))]
```

```python
plt.plot(range(199), arr, color='r')
plt.plot(range(199), arr1, color='g')

plt.show()
```
