```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pylab import *
from scipy.stats import *
from sklearn.metrics import mean_squared_error as msq
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
sample = norm.rvs(size=100)
alpha = 0.95
```

# a при известном σ2

```python
# Границы интервалов
# Положим σ2 = 1
def T1(sample):
    return mean(sample) - (norm.ppf((1+alpha)/2)/sqrt(len(sample)))

def T2(sample):
    return mean(sample) - (norm.ppf((1-alpha)/2)/sqrt(len(sample)))
```

```python
T_1 = []
T_2 = []
for i in range(1, 101):
    T_1 += [T1(sample[:i])]
    T_2 += [T2(sample[:i])]
plt.fill_between(range(1, 101), T_1, T_2)
plt.show()
```

# σ2 при известном a

```python
# Границы интервалов
def T1(sample):
    return 0

def T2(sample):
    return (sample ** 2).sum() / chi2.ppf(1-alpha, df=100)
```

```python
T_1 = []
T_2 = []
for i in range(1, 101):
    T_1 += [T1(sample[:i])]
    T_2 += [T2(sample[:i])]
plt.fill_between(range(1, 101), T_1, T_2)
plt.show()
```

# σ2 при неизвестном a

```python
# Границы интервалов
def T1(sample):
    return ((100 * var(sample)) / chi2.ppf(alpha/2, df=100))

def T2(sample):
    return ((100 * var(sample)) / chi2.ppf(1-(alpha/2), df=100))
```

```python
T_1 = []
T_2 = []
for i in range(1, 101):
    T_1 += [T1(sample[:i])]
    T_2 += [T2(sample[:i])]
plt.fill_between(range(1, 101), T_1, T_2)
plt.show()
```
